# Visualization and Explanatory Analysis

### 1. How's the performance of the sales a few months ago?

```python
import matplotlib.pyplot as plt
import seaborn as sns

all_df["order_month"] = all_df["order_date"].dt.month
all_df["order_year"] = all_df["order_date"].dt.year
plt.figure(figsize=(10, 5))
sns.lineplot(data=all_df, x="order_month", y="total_price", hue="order_year")
plt.ylabel("Total Sales")
plt.xlabel("Month")
plt.title("Total Sales per Month");
```

### 2. What Item Sales the Most and the least?

```python
product_sales = all_df.groupby(by="product_name").agg({
    "quantity_x": "sum",
    "total_price": "sum",
}).sort_values(by="quantity_x", ascending=False).head(10)

plt.figure(figsize=(12, 6))
colors = sns.color_palette('dark', len(product_sales))
sns.barplot(data=product_sales.reset_index(), x="product_name", y="quantity_x", palette=colors)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Total Sales")
plt.xlabel("Product Name")
plt.title("Top 10 Most Selling Products")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of each bar
for index, value in enumerate(product_sales["quantity_x"]):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.show()
```

```python
product_sales = all_df.groupby(by="product_name").agg({
    "quantity_x": "sum",
    "total_price": "sum",
}).sort_values(by="quantity_x", ascending=False).tail(10)

plt.figure(figsize=(12, 6))
colors = sns.color_palette('dark', len(product_sales))
sns.barplot(data=product_sales.reset_index(), x="product_name", y="quantity_x", palette=colors)
plt.xticks(rotation=45, ha='right')
plt.ylabel("Total Sales")
plt.xlabel("Product Name")
plt.title("Top 10 Least Selling Products")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of each bar
for index, value in enumerate(product_sales["quantity_x"]):
    plt.text(index, value, str(value), ha='center', va='bottom')
```

# Advanced Analysis

#### Recency

```python
recency = all_df.groupby(by="customer_id").order_date.max().reset_index()
recency.columns = ["customer_id", "last_purchase_date"]
recency["recency"] = (all_df["order_date"].max() - recency["last_purchase_date"]).dt.days
recency.drop(columns="last_purchase_date", inplace=True)
recency
```

#### Frequency

```python
frequency = all_df.groupby(by="customer_id").order_id.nunique().reset_index()
frequency.columns = ["customer_id", "frequency"]
frequency
```

#### Monetary

```python
monetary = all_df.groupby(by="customer_id").total_price.sum().reset_index()
monetary.columns = ["customer_id", "monetary"]
monetary
```

#### Merge and Visualize the Recency Distribution

```python
rfm = pd.merge(
    left=recency,
    right=frequency,
    how='left',
    left_on="customer_id",
    right_on="customer_id"
)

rfm = pd.merge(
    left=rfm,
    right=monetary,
    how='left',
    left_on="customer_id",
    right_on="customer_id"
)

plt.figure(figsize=(12, 6))
colors = sns.color_palette('pastel', 20)
ax = sns.histplot(data=rfm, x="recency", bins=20, kde=True, color=colors[0])
plt.ylabel("Total Sales")
plt.xlabel("Recency")
plt.title("Recency Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:  
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 9), 
                    textcoords='offset points')

plt.show()
```

# Cluster Analysis

#### Perform clustering analysis using manual grouping

```python
def rfm_cluster(row):
    if row['recency'] <= 30 and row['frequency'] >= 5 and row['monetary'] >= 1000:
        return 'Best Customers'
    elif row['recency'] <= 90 and row['frequency'] >= 3 and row['monetary'] >= 500:
        return 'Loyal Customers'
    elif row['recency'] <= 180 and row['frequency'] >= 1 and row['monetary'] >= 100:
        return 'Potential Loyalists'
    elif row['recency'] > 180 and row['frequency'] >= 1 and row['monetary'] >= 100:
        return 'At Risk'
    else:
        return 'Others'

rfm['cluster'] = rfm.apply(rfm_cluster, axis=1)
print(rfm['cluster'].value_counts())

plt.figure(figsize=(12, 6))
colors = sns.color_palette('pastel', len(rfm['cluster'].unique()))
ax = sns.countplot(data=rfm, x='cluster', palette=colors)
plt.title('Customer Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 9), 
                    textcoords='offset points')

plt.show()
```

```python
# perform cluster analysis with binning method
bins = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360]
labels = ['0-30', '31-60', '61-90', '91-120', '121-150', '151-180', '181-210', '211-240', '241-270', '271-300', '301-330', '331-360']
rfm['recency_bin'] = pd.cut(rfm['recency'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(12, 6))
colors = sns.color_palette('pastel', len(labels))
ax = sns.countplot(data=rfm, x='recency_bin', palette=colors)
plt.title('Customer Clusters by Recency Bins')
plt.xlabel('Recency Bin')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='center', 
                    xytext=(0, 9), 
                    textcoords='offset points')

plt.show()
```
